```python
W = dataset.iloc[:, 2:4].values 
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 1)
M = pca.fit_transform(W)
M
```

```python
from sklearn.preprocessing import StandardScaler
se =StandardScaler()
M=se.fit_transform(M)
```

```python
M
```

```python
dataset['PCA2']=M
```

```python
Z=dataset.iloc[:, 4:6].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,20):
  kmeans= KMeans(n_clusters = i, init = 'k-means++', random_state=42)
  kmeans.fit(Z)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,20),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
     
```

```python
kmeans= KMeans(n_clusters = 5, init = 'k-means++', random_state=40)
Y_predkmeans=kmeans.fit_predict(Z)
```

```python
print(Y_predkmeans)
```

```python
dataset['Cluster_label']=Y_predkmeans
```

```python
dataset.head()
```

```python

plt.scatter(Z[Y_predkmeans==0,0], Z[Y_predkmeans==0,1], s=100, c='red' , label='cluster 1')
plt.scatter(Z[Y_predkmeans==1,0], Z[Y_predkmeans==1,1], s=100, c='blue' , label='cluster 2')
plt.scatter(Z[Y_predkmeans==2,0], Z[Y_predkmeans==2,1], s=100, c='green' , label='cluster 3')
plt.scatter(Z[Y_predkmeans==3,0], Z[Y_predkmeans==3,1], s=100, c='cyan' , label='cluster 4')
plt.scatter(Z[Y_predkmeans==4,0], Z[Y_predkmeans==4,1], s=100, c='magenta' , label='cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], s=300, c='yellow', label='Centroids' )
plt.title('Cluster of Questions')
plt.xlabel('PCA_procc')
plt.ylabel('PCA2')
plt.legend(loc=4)
plt.show()
```

```python
dataset3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Research/r3.csv')
XX = dataset3.iloc[:,:5].values
YY = dataset3.iloc[:,5].values
ZZ = dataset3.iloc[:,0:6].values
```

```python
XX_train,XX_test,YY_train,YY_test= train_test_split(XX,YY, test_size=0.3,random_state=0)
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
n_input = 10
n_features = 6
generator = TimeseriesGenerator(ZZ,ZZ,length=n_input, batch_size=1)
```

```python
XXXX,YYYY=generator[0]
```

```python
XXXX
```

```python
YYYY
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
import tensorflow as tf
```

```python
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(generator,epochs=50)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```
