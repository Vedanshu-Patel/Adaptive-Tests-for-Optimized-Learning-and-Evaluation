```python
Z=dataset.iloc[:, 4:6].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,20):
  kmeans= KMeans(n_clusters = i, init = 'k-means++', random_state=42)
  kmeans.fit(Z)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,20),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
     
```

```python
kmeans= KMeans(n_clusters = 5, init = 'k-means++', random_state=40)
Y_predkmeans=kmeans.fit_predict(Z)
```

```python
print(Y_predkmeans)
```

```python
dataset['Cluster_label']=Y_predkmeans
```

```python
dataset.head()
```

```python

plt.scatter(Z[Y_predkmeans==0,0], Z[Y_predkmeans==0,1], s=100, c='red' , label='cluster 1')
plt.scatter(Z[Y_predkmeans==1,0], Z[Y_predkmeans==1,1], s=100, c='blue' , label='cluster 2')
plt.scatter(Z[Y_predkmeans==2,0], Z[Y_predkmeans==2,1], s=100, c='green' , label='cluster 3')
plt.scatter(Z[Y_predkmeans==3,0], Z[Y_predkmeans==3,1], s=100, c='cyan' , label='cluster 4')
plt.scatter(Z[Y_predkmeans==4,0], Z[Y_predkmeans==4,1], s=100, c='magenta' , label='cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], s=300, c='yellow', label='Centroids' )
plt.title('Cluster of Questions')
plt.xlabel('PCA_procc')
plt.ylabel('PCA2')
plt.legend(loc=4)
plt.show()
```

```python
dataset3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Research/r3.csv')
XX = dataset3.iloc[:,:5].values
YY = dataset3.iloc[:,5].values
```

```python
XX_train,XX_test,YY_train,YY_test= train_test_split(XX,YY, test_size=0.3,random_state=0)
```

```python
from xgboost import XGBRegressor
my_model = XGBRegressor()
my_model.fit(XX_train, YY_train, verbose=False)
```

```python
predictions = my_model.predict(XX_test)
for i in range(len(predictions)):
  predictions[i] = round(predictions[i])
from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error : " + str(mean_absolute_error(predictions, YY_test)))
```

```python
my_model = XGBRegressor(n_estimators=1000)
my_model.fit(XX_train, YY_train, early_stopping_rounds=5, 
             eval_set=[(XX_test, YY_test)], verbose=False)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of Model::",accuracy_score(YY_test,predictions))
```

```python
score = 3
numberOfQuestions = int(input('Enter the number of questions: '))
prevQuestions=0
penalty=0.00
for i in range(numberOfQuestions):
  time = int(input('Time taken for the question: '))
  result = int(input('The result: '))
  prevQuestions= prevQuestions +1
  penalty = penalty - 0.1*result*(time + prevQuestions)
  CC = [score,result,time,prevQuestions,penalty]
  RR = np.round(my_model.predict(CC)) 
  II = RR[0]
  score = II
  
  newRow = {'cluster':score,
            'result': result,
            'time' : time,
            'noq': prevQuestions,
            'penalties': penalty,
            'Score':II
            }
  dataset3 = dataset3.append(newRow,ignore_index=True)
  XXVV = dataset3.iloc[:,:5].values
  YYVV = dataset3.iloc[:,5].values

  VVmodel = XGBRegressor()
  XXVV_train,XXVV_test,YYVV_train,YYVV_test= train_test_split(XXVV,YYVV, test_size=0.3,random_state=0)
  VVmodel.fit(XXVV_train, YYVV_train, verbose=False)
  VVmodel = XGBRegressor(n_estimators=1000)
  VVmodel.fit(XXVV_train, YYVV_train, early_stopping_rounds=5,eval_set=[(XXVV_test, YYVV_test)], verbose=False)
```
